# **Assignment 2**

## Background
Senior financial analysts at a bank have some concerns about the credit risk models your (peer-to-peer lending) company has been using. They have based their concerns on performance benchmark figures that they generated using some historic lending data from our company. They conclude that these models are too ad-hoc to be suitable for use in a bank that is subject to strict regulatory requirements.

## Task
To rebuild the credit risk model from the ground up.

### Objective
To build a statistical model to **predict loan default based on information known at the time of the loan application**

# Key Questions

## Basic
1. How does the new model perform compared to the previous one?
    1. How can it be expected to perform on new loan applications?
2. What are the important variables in this model?  
    1. How do they compare to variables that are traditionally important for predicting credit risk in the banking sector?
        1. One regulatory requirement is that lenders need to be able to clearly explain how a loan application was assessed.
    2. Clearly interpret all covariates in your model in terms of their effect on predicting credit risk.

## Extended
3. Can accounting for this variation (E.g., state/zip code and time) improve performance benchmarks?
4. Are there any suprising differences in variables that are important for predicting credit risk?  
    1. this is also essential for regulations.
5. Does credit risk change over time or between states?
    1. This is not something the bank has previously investigated and results may inform modified loan policies in the future.
